```python

def categorize_loan(input_term):
    categories = {
        "Home Loans": [
            "home loan", "home buying", "home improvement", "home", "pool", "bathroom",
            "nri home loans", "loan against property",
            "home improvement loan"
        ],    
        "Personal Loans": [
            "personal", "personal loan",
            "Loan", "Loan1", "Get debt free",
            "Pay off", "My Loan"
        ],
        "Debt Consolidation": [
            "debt consolidation", "bill consolidation",
            "debt free", "get out of debt", "debt reduction",
            "Bill Payoff", 
            "Debt Payoff", "Consolidation Loan", "debt loan", "Freedom", "debt",
             "Debt Consolidation Loan",
        ],
        "Credit Card Management": [
            "credit cards", "cc", "credit card payoff",
            "credit card refinance", "credit payoff", "credit cards","credit","card",
            "credit card refinance loan","credit card debt", "credit card loan",
            "Credit consolidation", "CC refinance", "CC consolidation"
            
        ],
        "Vehicle Loans": [
            "car loan", "car financing", "vehicle loans", "major purchase", 
            
        ],
        "Business Loans": [
            "business", "working capital loans", "small industry", "trade credit", 
            "lending club", "lending loan"
        ],
        "Education Loans": [
            "education loans"
        ],
        "Secured Loans": [
            "loan against property", "loan against securities", "gold loan", "loan against ppf"
        ],
        "Special Purpose Loans": [
            "wedding loan", "vacation", "travel loans", "medical expenses", "medical loan",
            "moving and relocation", "consumer durable loans"
        ],
        "Agriculture Loans": [
            "agriculture","green loans"
        ],
        "Other Financial Products": [
            "overdraft", "cash advances", "title loans", "Bills","Lending Club", "My Loan" ,"Pay Off","Refinance","Refinance Loan"
        ]
    }

    input_term = input_term.lower()
    for category, terms in categories.items():
        if any(term in input_term for term in terms):
            return category
    
    return "Uncategorized"

def main():
    print("Welcome to the Loan Category Classifier!")
    print("Enter a loan term, and I'll categorize it for you.")
    print("Type 'quit' to exit the program.")
    print()

    while True:
        user_input = input("Enter a loan term: ")
        if user_input.lower() == 'quit':
            print("Thank you for using the Loan Category Classifier. Goodbye!")
            break
        category = categorize_loan(user_input)
        print(f"Category: {category}")
        print()

if __name__ == "__main__":
    main()

set(DF['Loan Title']) 
```
